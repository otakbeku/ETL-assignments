```python
def pandas_to_sqlite(data:pd.DataFrame, table_name:str, db_name:str, if_exists="replace", index=False) -> None:
    conn = sqlite3.connect(db_name)
    c = conn.cursor()
    CREATE_TABLE_QUERY = '''CREATE TABLE IF NOT EXISTS'''
    value_query = f" {table_name} ("
    for cl, dtype in zip(data.columns, data.dtypes):
        cl = cl.replace('-', '_')
        if dtype == 'int64' or dtype == 'float64':
            dtype = 'number'
        else:
            dtype = 'text'
        value_query += f"{cl} {dtype} "
    value_query += ")"
    c.execute(CREATE_TABLE_QUERY + value_query)
    data.to_sql(table_name, conn, if_exists=if_exists, index=index)
    conn.commit()
    conn.close()
    
```

```python
pandas_to_sqlite(df_final, "Final_superstore", "trial.db")
```

```python
pandas_to_sqlite(df_product, "Product_superstore", "trial.db")
```
